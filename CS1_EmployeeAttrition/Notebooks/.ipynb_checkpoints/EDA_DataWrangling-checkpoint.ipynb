# Attrition Vs Features

I'd like to see the rate of Attrition for differnet features
I groupby feature to get the Attrition rate and then pivot the dataframe to plot easily

```python
def get_group_rates(df,col,target,drop_values=None):
        grouped = df.groupby([col])[target].value_counts(normalize=True).reset_index(name='Rate') #Get rates for grouping
        pivot = grouped.pivot_table(index=col,columns=target,values='Rate') #Pivot for easy graphing
        if  drop_values != None:   
            pivot = pivot.drop(drop_values,axis=1) #drop any rates we don't want to see (for instance reduntant %s (Yes vs NO))
        return pivot
    
def plot_group(pivot):
            plt.clf()
            plt.figure(figsize=(15,15))
            pivot.plot(kind='bar',stacked=True)
            plt.xticks(rotation=65)
            plt.show()

def plot_rate(df,col,target,drop_values=None): #just an easier abstraction
    plot_group(get_group_rates(df,col,target,drop_values))
    
            

def plot_col_rates(df,target,drop_values=None,columns_to_drop=columns_to_drop):
    
    binned_cols = list(col[:-3] for col in df.columns if col.endswith('Bin'))
    cols = list(col for col in df.columns if (col not in columns_to_drop) and (col not in binned_cols)) #onl do non-dedundant columns

    for col in cols:
        if col != target:
            plot_rate(df,col,target,drop_values)
```

Some of the features have may unique values so I'll "bin" them to make teh graph easier to see

```python
bins = {'Age':10,'HourlyRate':10,'MonthlyRate':500,"MonthlyIncome":500,"DailyRate":100,
       'DaysWorked':5,"MonthlyHours":20,
        'CompanyRatiotoCareer':.05,'RoleRatioToCompany':.05,'RoleRatioToCareer':.05}

for col,y in bins.items():
    df[col+'Bin'] = df[col].apply(lambda x:x - x%y)
```

```python
plot_col_rates(df,'Attrition',drop_values='No')
```
