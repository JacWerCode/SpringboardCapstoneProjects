# Data Wrangling

## Raw Data

The original data was in a  sav file that caused problems when ding text manipulation, so it is immediately saved as a csv file and reloaded to solve the issue. Just about all the columsn ended with "W50", as this is part of the Pew Reasearch American Trends Wave 50 Survey. "F" was also used to designate non-survey questions.

```python
def get_raw_df(rename=False):
    sav_df = pd.read_spss('../W50_Jun19/ATP W50.sav')
    sav_df.to_csv('../Data/0_RAW.csv',index=False)
    df = pd.read_csv('../Data/0_RAW.csv',index_col='QKEY')
    df.columns = [col.replace("_W50","") for col in df.columns] # Many of the Questions originally end in _W50
    df.columns = [col.replace("F_","") for col in df.columns]
    return df
```

```python
raw_df = get_raw_df()
print(raw_df.shape)
raw_df.head(3)
```

## Converting  Questions to numerics

Much of the data is string data indicating a strength of agreement / satisfaction/ importance that can be converted into an integer. I created Replacements.csv common answers and their weights

```python
replacements = pd.read_csv('../Data/Replacements.csv',index_col='Entry')
replacements.head(8)
```

```python
def getNumericDf(df,verbose=False):
    numeric_df = df.copy()
    replacements = pd.read_csv('../Data/Replacements.csv',index_col='Entry').to_dict()['Value']
    
    change = {f'^{k}$':v for k,v in replacements.items()}
    numeric_df.replace(regex = change,inplace=True)
    for col in numeric_df.columns:
        try:
            numeric_df[col] = numeric_df[col].astype('float')
        except:
            pass
    return numeric_df


numeric_df = getNumericDf(raw_df,verbose=False)
numeric_df.to_csv('../Data/1_Numeric.csv')
numeric_df.head(3)
```

The names of the columns are not very informative so I use another file Question_Key.csv to map some more informative names to the features. Tabs and spaces where used to make the csv easier to read in the notepad and are stripped.

```python
def rename_cols(df):
    col_keys_df = pd.read_csv('../Data/Question_Key.csv',index_col='Key')
    col_keys_df['Value'] = col_keys_df['Value'].str.strip("\t")
    col_keys_df['Dependencies'] = col_keys_df['Dependencies'].str.strip("\t")
    
    col_keys = col_keys_df.to_dict()['Value']

    return df.rename(columns=col_keys)

numeric_df = pd.read_csv('../Data/1_Numeric.csv',index_col='QKEY')
renamed_df = rename_cols(numeric_df)
renamed_df.to_csv('../Data/2_Renamed.csv')
renamed_df.head(2)
```

```python
def impute_values(df):

    new_df=df.copy()
    '''

    object_cols = new_df.select_dtypes('object')
    for col in object_cols:
        new_df[col].fillna('Refused',inplace=True)
    '''
    return new_df


renamed_df = pd.read_csv('../Data/2_Renamed.csv',index_col='QKEY')
imputed_df = impute_values(renamed_df)
imputed_df.to_csv('../Data/3_Imputed_df.csv')
```

```python
def engineer_features(df):
    new_df=df.copy()

    
    for sex in ['Man','Woman']:
        new_df[f'{sex}FulfillmentCommitment'] = new_df[f'{sex}FulfillmentCommitment'].fillna(new_df.pop(f'{sex}FulfillmentMarriage'))

    for reason in ['Love','Companionship','Pregnancy','Children','FinancialReasons','Convenience','FormalCommitment']:
        new_df[f'For{reason}'] = new_df[f'CohabitatedFor{reason}'].fillna(new_df.pop(f'MarriedFor{reason}'))
        new_df.drop([f'CohabitatedFor{reason}'],axis=1,inplace=True)
    
    for prereq in ['FinancialStability','SteadyJob','House','CompleteCommitment']:
        new_df[f'{prereq}Before'] = new_df[f'{prereq}BeforeCohabitation'].fillna(new_df.pop(f'{prereq}BeforeMarriage'))
        new_df.drop([f'{prereq}BeforeCohabitation'],axis=1,inplace=True)
    

#    new_df['INCOME'] = new_df['INCOME'].apply(lambda x: re.findall(r"\$(\d{2,3}),",x))
 #   new_df['INCOME'] = new_df['INCOME'].apply(lambda x: [0,10] if x == ['10'] else x)
  #  new_df['INCOME'] = new_df['INCOME'].apply(lambda x: int(x[0])*1000 if len(x)>0 else np.nan)
   

    
#    for col in new_df.select_dtypes('object').columns:
#        print(col)
#        for p in ['<',">",",","-","/","'"," "]:
#            new_df[col] = new_df[col].str.strip(p) 
                
    return new_df
imputed_df = pd.read_csv('../Data/3_Imputed_df.csv',index_col='QKEY')
clean_df = engineer_features(imputed_df)
clean_df.to_csv('../Data/4_Clean.csv')
clean_df.shape
```


Data was gained from an online survey on mobile or desktop. The survey was programmed to only ask "relevant" questions based on past answers, (example "Are you currently engaged" would only be asked to unmarried participants). In the same Question_Key.csv file I have entred the dependancies of each question and here will only use questions that were asked to everyone, or are relativent to our model.


```python
def prune_cols(df):
    # get columns to use based on dependancies
    col_keys_df = pd.read_csv('../Data/Question_Key.csv',index_col='Key')
    col_keys_df['Value'] = col_keys_df['Value'].str.strip("\t")
    col_keys_df['Value'] = col_keys_df['Value'].str.strip(" ")
    col_keys_df['Dependencies'] = col_keys_df['Dependencies'].str.strip("\t")
    
    dependencies = ['BASE','FORM=2','FAMSURV12=1','MARITALMOD=1-4']
    cols_by_dependancy  = col_keys_df[col_keys_df['Dependencies'].isin(dependencies)]['Value']

    #List of engineered columns that would otehrwise not be included frm dependancies
    before_together =  [f'{prereq}Before' for prereq in ['FinancialStability','SteadyJob','House','CompleteCommitment']]
    
    
    mask_cols = list(set(cols_by_dependancy).intersection(set(clean_df.columns))) + before_together
    
    drop_cols = ['ACSWEB','VOLSUM','INCOME_RECODE','BORN','FORM','SocietyBetterMarried']
    
    return df[mask_cols].drop(drop_cols,axis=1)

clean_df = pd.read_csv('../Data/4_Clean.csv',index_col='QKEY')
final_df = prune_cols(clean_df)
final_df.to_csv('../Data/5_Final.csv')
final_df.shape
```

# EDA

## Helper functions

```python
def plot_heat_map(df,x,y,value,cmap='Blues',aggfuncs =['count','mean'],annot=True,T=False):
    for aggfunc in aggfuncs:
        if aggfunc == "count":
            fmt = '.0f'
            plt.title(f"Counts by {x} and {y}")
        else:
            fmt = '.2f'
            plt.title(f"{aggfunc.capitalize()} {value} by {x} and {y}")
            
            
        heat_df = pd.pivot_table(data=df,columns=x,index=y,values=value,aggfunc=aggfunc).sort_index(ascending=False)
        ignore = 'Refused'
        refused_col = [col for col in heat_df.columns if ignore in col]
        refused_ind = [ind for ind in heat_df.index if ignore in ind]
        heat_df.drop(refused_col,axis=1,inplace=True)
        heat_df.drop(refused_ind,axis=0,inplace=True)
        if T:
            heat_df = heat_df.T
            
        sns.heatmap(heat_df,cmap=cmap,annot=annot,fmt=fmt,cbar=False,square=False,linecolor='Black',linewidths=1)
        plt.title(f'{aggfunc} by {x} and {y}')
        plt.show()
        
    return heat_df
```

```python
def graph_groupby(df,x,y,y_names,cmap='Blues',annot=True,T=False,rot=0,cbar=False,):
    
    heat_df = df.groupby(x)[y].mean()
    
    ignore = 'Refused'
    
    refused_col = [col for col in heat_df.columns if ignore in col]
    refused_ind = [ind for ind in heat_df.index if ignore in ind]
    heat_df.drop(refused_col,axis=1,inplace=True)
    heat_df.drop(refused_ind,axis=0,inplace=True)
    
    
    if T:
        heat_df = heat_df.T
    
    sns.heatmap(heat_df,cmap=cmap,annot=annot,fmt='.2f',cbar=cbar,square=False,linecolor='Black',linewidths=0)
    plt.title(f'Mean {y_names} by {"-".join(x)}')
    
    plt.xticks(rotation=rot)
    return heat_df
```
